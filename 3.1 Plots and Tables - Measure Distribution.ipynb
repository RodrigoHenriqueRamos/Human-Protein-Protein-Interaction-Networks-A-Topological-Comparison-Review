```python
import warnings
warnings.filterwarnings('ignore')
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read measures

```python
measures={}
for name in ['HINT','IntAct','Reactome','STRING']:
    measures[name]=pd.read_csv('3 - output/resume_'+name+'.tsv',sep='\t',index_col=0)
```

### Top Measures

```python
for measure in ['Clustering','Degree','AvgNeighbors','PageRank','Eigenvector']:
    measureName=measure
    plt.figure(figsize=(17,4),dpi=200)
    for i,name in enumerate(sorted(measures)):
        plt.subplot(1,4,i+1)
        values=sorted(measures[name][measure])
        plt.plot(values,'.')
        plt.title(name+' - Avg:'+str(round(np.mean(measures[name][measure]),2)),fontsize=18)  
        if i == 0:
            plt.ylabel(measureName,fontsize=16)  
        #plt.xlabel('Nodes',fontsize=16)
        plt.yticks(fontsize=16)
        plt.xticks([0,
                    round(len(values)*0.25),
                    round(len(values)*0.5),
                    round(len(values)*0.75),
                    len(values)
                   ],fontsize=14,rotation=0)
        plt.tight_layout()
        
#    plt.suptitle(measureName,fontsize=24)
    plt.tight_layout()
    plt.savefig("3.1 - output/"+measureName+".png", format='png', dpi=550)
    plt.savefig("3.1 - output/"+measureName+".pdf", format='pdf')
    plt.show()
    break
```
